```python
import requests
from bs4 import BeautifulSoup

# send a request to the NBC News homepage and get the HTML content
response = requests.get("https://www.nbcnews.com/")
html_content = response.content

# create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# find all the text on the page
all_text = []
for element in soup.find_all():
    if element.name == 'script' or element.name == 'style' or element.name == 'meta':
        continue
    text = element.get_text().strip()
    if len(text) > 0:
        all_text.append(text)

# print the extracted text
print('\n'.join(all_text))

# find all the links on the page
all_links = []
for link in soup.find_all('a'):
    href = link.get('href')
    text = link.get_text().strip()
    if len(href) > 0 and len(text) > 0:
        all_links.append((text, href))

# print the extracted links and their text
for link in all_links:
    print(link[0], link[1])
    # send a request to the link and get the HTML content
    try:
        response = requests.get(link[1], timeout=5)
        html_content = response.content
        # parse the HTML content with BeautifulSoup to extract the text
        soup = BeautifulSoup(html_content, 'html.parser')
        link_text = ' '.join(soup.stripped_strings)
        print(link_text)
    except:
        print("Error: Could not retrieve link content")
    print('')
```
